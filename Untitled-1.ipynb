```python
import re
# converting text from proper case into lower case
df['content'] = df['content'].apply(lambda x : x.lower())
df.head()
```

converting strings in category variable into integers to facilitate cleaning of data

```python
targetmapping = {'travel':0, 'art_music':1, 'history': 2, 'food': 3}
df['category'] = df['category'].apply(lambda x : targetmapping[x])
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def preprocessing(data):
    corpus = []
    for i in range(len(data)):
        # removing urls
        text1 = re.sub(r'https\S+', ' ', data['content'][i])
        #removing digits
        text3 = re.sub(r'\d+', ' ', text1)
        # removing hashtags
        text4 = re.sub(r'#\w+', ' ', text3)
        # removing punctuation marks
        text = re.sub(r'[^a-zA-Z]', ' ', text4)
        # creating spaces in the text
        text = text.split()
        # using stopwords to remove words such as aticles, auxilliary verbs and preposition
        text = [word for word in text if word not in stopwords.words('english')]
        # applying portestemmer to reduce words into their basic form
        text = [ps.stem(word) for word in text]
        text = ' '.join(text)
        corpus.append(text)
    return corpus
```

```python
# applying the function to clean the dataset
corpus = preprocessing(df)
```

```python
# a sample of the cleaned data
corpus[:10]
```

```python
# aplying TYfid vectorizer which gives maximum weight to important words in the text by checking the frequency of a word in a
# number of documents
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
articles = tfid.fit_transform(corpus)
```

Kmeans clustering is an unsupervised machine learning algorithm that groups data inform of clusters which has the same properties.it uses an iterative process by finding the appropriate k (number of iterations) which are necessary to group data into the specified clusters.
TruncatedSVd is used to reduce dimensionality that is it replaces redudant features with appropriate new features that have a summarized information of the original feature space.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD()
arrticles = svd.fit_transform(articles)
```

```python
#creating an empty list
# finding the most appropriate number of clusters to feed the model
ssd = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i)
    km = kmeans.fit(arrticles)
    ssd.append(kmeans.inertia_)
    labels = kmeans.labels_
    
```

```python
# a plot of the range of values for the number of clusters
plt.figure(figsize=(10, 8))
plt.title('elbow plot for visualising appropriate i clusters')
plt.plot(range(2, 10), ssd, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('ssd')
plt.show()
```

an elbow is formed at the 4th value thus that is the appopriate value for n_clusters.

```python
# conducting clustering using 4 as the number of clusters
kmeans = KMeans(n_clusters=4)
clusters = kmeans.fit_predict(arrticles)
labels = kmeans.labels_
centres = kmeans.cluster_centers_
labels = kmeans.labels_
#arrticles['cluster_labels'] = labels
```

```python
# applying TSNE which is used to display high dimensionational data
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
ts = tsne.fit_transform(arrticles)
tsne_df = pd.DataFrame(data=ts, columns=['temp1', 'temp2'])
tsne_df.head()
```

```python
# visualising clusters in the dataset
plt.figure(figsize=(10, 10))
sns.scatterplot(x=tsne_df.iloc[:, 0], y=tsne_df.iloc[:, 1], cmap='brg', c=clusters, label=labels)
plt.ylabel('temp1')
plt.xlabel('temp2')
plt.title('clustering of the arrticles')
plt.legend()
plt.show()
```

splitting the cleaned data into x and y variables.This will facilitate the prediction of the data.

```python
# splitting the data into x and y variables
x = corpus
y = df['category']
```

```python
# splitting x and y variables into training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20) 
x_train = tfid.fit_transform(x_train)
x_test = tfid.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
acc = lr.score(x_test, y_test)
pred = lr.predict(x_test)
print(acc)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(pred, y_test))
```
